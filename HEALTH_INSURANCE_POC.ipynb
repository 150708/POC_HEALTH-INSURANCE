```python
#data cleaning

df =  df.drop(['ID','City_Code','Region_Code','Accomodation_Type','Is_Spouse','Holding_Policy_Duration','Holding_Policy_Type','Health Indicator'],axis = 1)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
#Label encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Reco_Insurance_Type'] = le.fit_transform(df['Reco_Insurance_Type'])
```

```python
#x,y split
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
#train_test_split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.25,random_state = 13)
```

```python
x_train
```

```python
y_train
```

```python
#standardizing
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train_sc = sc.fit_transform(x_train)
x_test_sc = sc.transform(x_test)

```

```python
#building a Support vector classifier model
from sklearn.svm import SVC
svc = SVC(kernel = 'linear')
svc.fit(x_train_sc,y_train)
```

```python
y_pred = svc.predict(x_test_sc)
```

```python
#evaluation metrics
from sklearn.metrics import confusion_matrix,accuracy_score,f1_score,precision_score,recall_score
cm = confusion_matrix(y_test,y_pred)
print("cm",cm)
```

```python
acc = accuracy_score(y_test,y_pred)
pre = precision_score(y_test,y_pred)
recall = recall_score(y_test,y_pred)
f1 = f1_score(y_test,y_pred)
```

```python
print("accuracy",acc)
print("precision",pre)
print("recall",recall)
print("f1",f1)
```
